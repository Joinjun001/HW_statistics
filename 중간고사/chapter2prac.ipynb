```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, uniform, norm, chi2, t, f              
```

```python
## HW2 

n = 500

result = np.random.choice([1,0], size=n, replace=True)
result[:50]
```

```python
trial = np.arange(1, n+1) # 각 동전던질때까지의 시행횟수
cumul_sum = np.cumsum(result) 
cumul_sum[450:499]
```

```python
prob = cumul_sum / trial
```

```python
plt.plot(trial,prob,marker="o",linestyle='-',markersize=5)
plt.ylim(-0.1,1.1)
plt.xlabel("trial")
plt.ylabel("probability")
plt.axhline(y=0.5,color="red",linewidth=1)
plt.show()
```
